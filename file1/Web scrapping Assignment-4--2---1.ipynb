```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
import warnings
from selenium.common.exceptions import NoSuchElementException
warnings.filterwarnings("ignore")
import time
```

```python
class getWebDriver:
    def __init__(self):
        chrome_options = webdriver.ChromeOptions()
        chrome_options.add_experimental_option("detach", True)
        self.driver = webdriver.Chrome(options=chrome_options)
        
    def makeDriver(self,url):
        self.driver.get(url)
        return self.driver
```

## 1
Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views 

```python
driver_object = getWebDriver()
driver = driver_object.makeDriver("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

time.sleep(3)

data = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr')
print(len(data))

raw_data = []
```

```python
for item in range(len(data)):
    inner_list = []
    get_videoname = data[item].find_element(By.XPATH,'./td[1]')
    get_uploadername = data[item].find_element(By.XPATH,'./td[2]')
    get_views = data[item].find_element(By.XPATH,'./td[3]')
    get_date = data[item].find_element(By.XPATH,'./td[4]')
    inner_list.append(item)
    inner_list.append(get_videoname.text)
    inner_list.append(get_uploadername.text)
    inner_list.append(get_views.text)
    inner_list.append(get_date.text)
    raw_data.append(inner_list)
        
```

```python
df = pd.DataFrame(raw_data,columns=['Rank','VideoName','UploadedBy','View (in Billions)','Upload Date'])
df.set_index('Rank', inplace=True)
df['VideoName'] = df['VideoName'].str.replace('"', '')
df
```

## 4
Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used 
Note: - From the home page you have to click on the trending option from Explore menu through code

```python
driver_object = getWebDriver()
driver = driver_object.makeDriver("https://github.com/")

time.sleep(1)

open_source = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
action_chains = ActionChains(driver)
action_chains.move_to_element(open_source).perform()

trending_click = driver.find_element(By.XPATH, '/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trending_click.click()

time.sleep(3)
```

```python
data = driver.find_elements(By.XPATH, '/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article')
raw_data = []

for item in range(len(data)):
    inner_list = []
    get_name = data[item].find_element(By.XPATH,'./h2/a/span')
    get_name2 = data[item].find_element(By.XPATH,'./h2/a')
    name = get_name.text + get_name2.text
    try:
        get_desc = data[item].find_element(By.XPATH,'./p').text
    except:
        get_desc = '---'
        
    get_count = data[item].find_element(By.XPATH,'./div[2]/a[1]').text
    try:
        get_lang = data[item].find_element(By.XPATH,'./div[2]/span/span[2]').text
    except:
        get_lang = '---'
    
    inner_list.append(name)
    inner_list.append(get_desc)
    inner_list.append(get_count)
    inner_list.append(get_lang)
    raw_data.append(inner_list)
    
```

```python
df = pd.DataFrame(raw_data,columns=['Repository title','Repository description','Contributors count','Language used'])
df
```

## 7
Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/ You have
to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes 

```python
driver_object = getWebDriver()
driver = driver_object.makeDriver("https://www.imdb.com")

menu_click = driver.find_element(By.XPATH, '/html/body/div[2]/nav/div[2]/label[1]/span')
menu_click.click()

time.sleep(1)

yes_click = driver.find_element(By.XPATH, '/html/body/div[2]/nav/div[2]/aside[1]/div/div[2]/div/div[2]/div[1]/span/div/div/ul/a[2]/span')
yes_click.click()

time.sleep(3)
raw_data = []
```

```python
data = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div/div[3]/section/div/div[2]/div/ul/li')

for item in range(len(data)):
    inner_list = []
    get_name = data[item].find_element(By.XPATH,'./div[2]/div/div/div/a/h3').text
    get_Year = data[item].find_element(By.XPATH,'./div[2]/div/div/div[2]/span[1]').text
    Run = data[item].find_element(By.XPATH,'./div[2]/div/div/div[2]/span[2]').text
    Ratings = data[item].find_element(By.XPATH,'./div[2]/div/div/span/div/span').text
    
    inner_list.append(get_name)
    inner_list.append(get_Year)
    inner_list.append(Run)
    inner_list.append(Ratings)

    raw_data.append(inner_list)
```

```python
df = pd.DataFrame(raw_data,columns=['Name','Year span','Run time','Ratings'])
df['Voting'] = df['Ratings'].str.split('\n').str[1]
df['Rating'] = df['Ratings'].str.split('\n').str[0]
del df['Ratings']
df['Series Name'] = df['Name'].str.split('.').str[1]
del df['Name']
df.insert(0,'Name', df['Series Name'])
del df['Series Name']

df['Voting'] = df['Voting'].str.replace("(", "").str.replace(")", "")
df
```
