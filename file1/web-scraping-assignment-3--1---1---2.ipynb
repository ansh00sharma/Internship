```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.webdriver.common.action_chains import ActionChains
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
class getWebDriver:
    def __init__(self):
        chrome_options = webdriver.ChromeOptions()
        chrome_options.add_experimental_option("detach", True)
        self.driver = webdriver.Chrome(options=chrome_options)
        
    def makeDriver(self,url):
        self.driver.get(url)
        return self.driver
```

### 1. 
Write a python program which searches all the product under a particular product from www.amazon.in. The
product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for
guitars. 

```python
user_input = input("Enter what you want to search : ")
driver_object = getWebDriver()
driver = driver_object.makeDriver("https://www.amazon.in/")

time.sleep(7)

search_bar = driver.find_element(By.XPATH, '/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input')
search_bar.send_keys(user_input)

click_search = driver.find_element(By.XPATH, '/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
click_search.click()
```

```python
name = []
price = []
exchange = []
delivery = []
availability = []
url = []

for i in range(0,2):
    name_element = driver.find_elements(By.XPATH, '//span[@class="a-size-base-plus a-color-base a-text-normal"]')
    url_element = driver.find_elements(By.XPATH, '//div[@class="a-section a-spacing-base"]/div[2]/div[1]/h2/a')
    price_element = driver.find_elements(By.XPATH,'//span[@class="a-price"]/span/span[2]')
    delivery_element = driver.find_elements(By.XPATH, '//div[@class="a-row s-align-children-center"]/span[2]/span[2]')
    
    for item in range(len(name_element)):
        
        name.append(name_element[item].text)
        try:

            url.append(url_element[item].get_attribute("href"))
        except:
            url.append('-')
        try:

            price.append(price_element[item].text)
        except:
            price.append('-')
        try:
            delivery.append(delivery_element[item].text)
            exchange.append('-')
            availability.append('-')
        except:
            delivery.append('-')
            exchange.append('-')
            availability.append('-')
            
        
    next_page = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    
    next_page.click()
    time.sleep(3)
```

### 3
Write a python program to access the search bar and search button on images.google.com and scrape 10
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’. 

```python
user_input = ['fruits','Guitar','cars','Machine Learning','Cakes']
data = {}
driver_object = getWebDriver()

for item in user_input:
    data[f'{item}'] = []
    driver = driver_object.makeDriver("https://images.google.com/")

    time.sleep(3) 
    # signout_click = driver.find_element(By.XPATH, '//div[@class="DRc6kd bdn4dc"]/div[2]/button')
    # signout_click.click()

    search_bar = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea')
    search_bar.send_keys(item)

    
    search_click = driver.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span')
    search_click.click()
    
    time.sleep(3)
    all_elements = driver.find_elements(By.XPATH, '//*[@id="islrg"]/div[1]/div/a[2]')
    for i in range(0,10):
        element = all_elements[i].get_attribute('href')
        data[f'{item}'].append(element)
        
    
```

```python
df = pd.DataFrame(columns=['topic','url'])

for keys in data:
    values = data[keys]
    
    for value in values:
        
        new_df = pd.DataFrame([[keys,value]])
        
        df = pd.concat([df, new_df], ignore_index=True)
        
        
df.drop(['topic','url'], axis=1,inplace=True)

# df.set_index('0',inplace=True)
df.rename(columns={0: "topic", 1: "url"}, inplace=True)
df
```

Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com

```python
driver_object = getWebDriver()
driver = driver_object.makeDriver("https://www.flipkart.com/")
```

```python
user_input = input("Enter what you want to search : ")
search_bar = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input')
search_bar.send_keys(user_input)

search_button = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button')
search_button.click()
```

```python
all_elements = driver.find_elements(By.XPATH, '//div[@class="_13oc-S"]')


names = []
specs = []
camera = []
battery = []
display = []
prices = []
warranty = []
url = []

for system in all_elements:
    item = system.find_element(By.XPATH,'./div/div/a/div[2]')
    name = item.find_element(By.TAG_NAME, 'div')
    info_1 = name.text.split('\n')    
    price = item.find_element(By.XPATH,'./div[2]')
    info_2 = price.text.split('\n')[0]
    info_3 = system.find_element(By.XPATH, './div/div/a')
    href = info_3.get_attribute('href')
    
    names.append(info_1[0])
    specs.append(info_1[2])
    display.append(info_1[3])
    camera.append(info_1[4])
    battery.append(info_1[5])
    warranty.append(info_1[-1])
    prices.append(info_2)
    url.append(href)
    
    
    
url
```

```python
df = pd.DataFrame(columns = ['full Name','storage','display','camera','battery','warranty','price','url'])
df['full Name'] = names
df['storage'] = specs
df['display'] = display
df['camera'] = camera
df['battery'] = battery
df['warranty'] = warranty
df['price'] = prices
df['url'] = url
df
```

```python
df['Brand'] = df['full Name'].str.split().str[0]
df['Name'] = df['full Name'].str.split('(').str[0]
df['Color'] = df['full Name'].str.split('(').str[1]
df['RAM'] = df['storage'].str.split('|').str[0]
df['ROM'] = df['storage'].str.split('|').str[1]
df['Primary Camera'] = df['camera'].str.split('|').str[0]
df['Secondary Camera'] = df['camera'].str.split('|').str[1].str.strip()
df['Secondary Camera'].fillna('-', inplace=True)
```

```python
del df['full Name']
del df['storage']
del df['camera']
df
```
