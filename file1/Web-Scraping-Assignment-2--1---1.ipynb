### Q1
Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:

1. First get the webpage https://www.shine.com/
2. Enter “Data Analyst” in “Job title, Skills” field and enter “Bangalore” in “enter the location” field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
driver = webdriver.Chrome()
driver.get("https://www.shine.com/")

click_search_bar = driver.find_element(By.XPATH, "/html/body/div/header/div[3]/div/div/div[1]/div/span")
click_search_bar.click()
```

```python
job_bot = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[1]/div/input")
job_bot.send_keys("Data Analyst")

location_bot = driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input")
location_bot.send_keys("Bangalore")

click_bot = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
click_bot.click()
```

```python
job_title = []
company_name = []
exerience = []
job_location = []
```

```python
result_titles = driver.find_elements(By.XPATH,'//div[@class="parentClass position-relative"]/div/div/div/h2')
for index in range(10):
    item = result_titles[index]
    title = item.text
    job_title.append(title)
    
result_company_name = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]/span')
for index in range(10):
    item = result_company_name[index]
    title = item.text
    company_name.append(title)
    
result_experience = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for index in range(10):
    item = result_experience[index]
    title = item.text
    exerience.append(title)
    
result_job_location = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for index in range(10):
    item = result_job_location[index]
    title = item.text
    job_location.append(title)
```

```python
data = pd.DataFrame(columns=["Job title","Company Name","Experience Required","Location"])
data['Job title'] = job_title
data['Company Name'] = company_name
data['Experience Required'] = exerience
data['Location'] =job_location
data
```

### Q4
Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. ProductDescription
3. Price
4. Discount

To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url :https://www.flipkart.com/
2. Enter “sunglasses” in the search fieldwhere “search for products, brands and more” is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
required data as usual.
4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100sunglasses.


```python
driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/")

click_search_bar = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/div/input")
click_search_bar.send_keys("sunglasses")

click_search_button = driver.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/div/div/div/div/div[1]/div/div[1]/div/div[1]/div[1]/header/div[1]/div[2]/form/div/button")
click_search_button.click()
```

```python
Brand = []
ProductDescription = []
Price = []
Discount = []
counter = 0
for i in range(3):
    
    result_Brand = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    result_ProductDescription = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    result_Price = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    result_Discount = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]/span')

    for index in range(len(result_Brand)):
        if(counter<100):
            try:
                b = result_Brand[index]
                des = result_ProductDescription[index]
                p = result_Price[index]
                d = result_Discount[index]

                Brand.append(b.text)
                ProductDescription.append(des.text)
                Price.append(p.text)
                Discount.append(d.text)

            except IndexError:
                b = result_Brand[index]
                p = result_Price[index]
                d = result_Discount[index]

                Brand.append(b.text)
                ProductDescription.append("N/A")
                Price.append(p.text)
                Discount.append(d.text)

            counter +=1
            
        else:
            break

    next_page_button = driver.find_element(By.XPATH, '//a[@class="ge-49M"]')
    next_page_button.click()
    driver.refresh()
    
```

```python
data = pd.DataFrame(columns=["Brand","ProductDescription","Price","Discount"])
data['Brand'] = Brand
data['ProductDescription'] = ProductDescription
data['Price'] = Price
data['Discount'] =Discount
data
```

### Q8
Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap 
a) Quote 
b) Author 
c) Type Of Quotes

```python
driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")

click_search_bar = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
click_search_bar.click()
```

```python
Quote = []
Author = []
Type = []

counter = 0
for i in range(10):
    result_Quote = driver.find_elements(By.XPATH,'//ul[@class="list-quotes"]/li/div/p/a[2]')
    result_Author = driver.find_elements(By.XPATH,'//ul[@class="list-quotes"]/li/div/div/a')
    result_Type = driver.find_elements(By.XPATH,'//div[@class="tags"]')

    for index in range(len(result_Quote)):
        if(counter<1000):
            print(counter)
            b = result_Quote[index]
            r = result_Author[index]
            p = result_Type[index]


            Quote.append(b.text)
            Author.append(r.text)

            type_texts = [tag.text for tag in p.find_elements(By.XPATH, './a')]
            Type.append(type_texts)
            
            counter +=1
    
    if(i == 0):
        next_page_button = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[12]/a')
    elif(i==9):
        break
    else: 
        next_page_button = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[4]/li[13]/a')
        
    next_page_button.click()
    print("next page clicked")
    driver.refresh()
```

```python
data = pd.DataFrame(columns=["Quote","Author","Type of Quote"])
data['Quote'] = Quote
data['Author'] = Author
data['Type of Quote'] = Type
data
```

```python
driver = webdriver.Chrome()
driver.get("https://www.motor1.com/")

features_section = driver.find_element(By.XPATH, '//*[@id="page_index_index_index"]/div[2]/div/div/div[3]/ul/li[3]/a')
features_section.click()
# //*[@id="page_index_index_index"]/div[2]/div/div/div[3]/ul/li[3]/a
```

```python
viewmore_section = driver.find_element(By.XPATH, "/html/body/div[10]/div[6]/div/div[1]/div[2]/a")
viewmore_section.click()

expensive_50_section = driver.find_element(By.XPATH, '//*[@id="page_index_articles_browse"]/div[8]/div[3]/div[1]/div/div/div[3]/div/div[1]/h3/a')
expensive_50_section.click()
```

```python
import re
Name = []
Price = []

result_Quote = driver.find_elements(By.XPATH,'//*[@id="article_box"]/div[1]/div[2]/div[2]/ul/li')
for index in result_Quote:
    sentence = index.text
    rev_sentence = sentence[::-1]
    pattern = r'[:,\-–]'
    data = re.split(pattern, rev_sentence, maxsplit=1)
    
    Name.append(data[0][::-1])
    Price.append(data[1][::-1])
   
        
      
```

```python
data = pd.DataFrame(columns=["Name","Price"])
data['Name'] = Price
data['Price'] = Name

data
```
